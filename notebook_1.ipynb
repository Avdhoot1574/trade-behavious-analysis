# Creating Dataframe for both Csv Files

```python
fg_df = pd.read_csv('fear_greed_index.csv', on_bad_lines='skip')
trader_df = pd.read_csv('trader_data.csv', on_bad_lines='skip')
display(trader_df.head())
```

# Overview Of Data of both Csv files

```python
display(fg_df.head())
display(trader_df.head())
```

# Correcting Timestamp IST format  for Normalizing

```python
trader_df['Timestamp IST'] = pd.to_datetime(trader_df['Timestamp IST'], format='%d-%m-%Y %H:%M', errors='coerce')
trader_df['date'] = trader_df['Timestamp IST'].dt.normalize()
```

```python
fg_df['date'] = pd.to_datetime(fg_df['date']).dt.normalize()
```

# Creating Merged Dataframe

```python
merged_df = pd.merge(trader_df, fg_df[['date', 'classification']], on='date', how='left')
print(merged_df['classification'].value_counts(dropna=False))
```

Bar plot for comparing long and short pnl by market sentiment

```python
account_pnl = merged_df.groupby('Account')['Closed PnL'].sum().sort_values(ascending=False).reset_index()
account_pnl['cumulative_pct'] = account_pnl['Closed PnL'].cumsum() / account_pnl['Closed PnL'].sum()

plt.figure(figsize=(10,6))
plt.plot(account_pnl.index, account_pnl['cumulative_pct'], color='purple')
plt.title('Pareto of Account PnL')
plt.xlabel('Accounts (sorted)')
plt.ylabel('Cumulative PnL %')
plt.grid(True)
plt.savefig('outputs/pareto_accounts.png')
plt.show()
```

# Kruskal- Wallis Test for statistical Validation
  the value of p is 0.000 then it is considered as most significant difference. it is very highly significant

```python
from scipy.stats import kruskal

groups = [group['Closed PnL'].values for name, group in merged_df.groupby('classification')]
stat, p = kruskal(*groups)
print(f"Kruskal-Wallis H-statistic: {stat:.2f}, p-value: {p:.4f}")
```

# Dunn's Test with scikit post-hoc and Bonferroni Correction
it showing the differences between the different sentiment combinations..

```python
!pip install scikit-posthocs
import scikit_posthocs as sp

posthoc_df = merged_df[['Closed PnL', 'classification']].dropna()

dunn_results = sp.posthoc_dunn(posthoc_df, val_col='Closed PnL', group_col='classification', p_adjust='bonferroni')
print(dunn_results)
```
